```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have already loaded or initialized your DataFrame (df)

# Function to visualize data and graphs
def visualize_data():
    plt.figure(figsize=(10, 6))
    plt.hist(df['MathScore'], color='orange', edgecolor='black')
    plt.title('Distribution of Maths Scores')
    plt.xlabel('Scores')
    plt.ylabel('Number of Students')
    plt.grid(True, alpha=0.3)
    plt.show()

    plt.figure(figsize=(10, 6))
    plt.hist(df['WritingScore'], color='Skyblue', edgecolor='black')
    plt.title('Distribution of Writing Scores')
    plt.xlabel('Scores')
    plt.ylabel('Number of Students')
    plt.grid(True, alpha=0.3)
    plt.show()

    plt.figure(figsize=(10, 6))
    plt.hist(df['ReadingScore'], color='yellow', edgecolor='black')
    plt.title('Distribution of Reading Scores')
    plt.xlabel('Scores')
    plt.ylabel('Number of Students')
    plt.grid(True, alpha=0.3)
    plt.show()

    plt.figure(figsize=(10, 6))
    ax = sns.countplot(data=df, x="Gender", palette="pastel")
    ax.bar_label(ax.containers[0])
    plt.title('Count of Students by Gender')
    plt.show()

    plt.figure(figsize=(10, 6))
    gb = df.groupby("Gender").agg({"MathScore": 'mean', "ReadingScore": 'mean', "WritingScore": 'mean'})
    sns.heatmap(gb, annot=True, cmap="YlGnBu")
    plt.title('Mean Scores by Gender')
    plt.show()

    plt.figure(figsize=(10, 6))
    gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore": 'mean', "ReadingScore": 'mean', "WritingScore": 'mean'})
    sns.heatmap(gb1, annot=True)
    plt.title('Mean Scores by Parent Marital Status')
    plt.show()

    plt.figure(figsize=(10, 6))
    gb2 = df.groupby("WklyStudyHours").agg({"MathScore": 'mean', "ReadingScore": 'mean', "WritingScore": 'mean'})
    sns.heatmap(gb2, annot=True)
    plt.title('Mean Scores by Weekly Study Hours')
    plt.show()

    plt.figure(figsize=(10, 6))
    gb3 = df.groupby("ParentEduc").agg({"MathScore": 'mean', "ReadingScore": 'mean', "WritingScore": 'mean'})
    sns.heatmap(gb3, annot=True)
    plt.title('Mean Scores by Parent Education Level')
    plt.show()

    plt.figure(figsize=(10, 6))
    gb = df.groupby("IsFirstChild").agg({"MathScore": 'mean', "ReadingScore": 'mean', "WritingScore": 'mean'})
    sns.heatmap(gb, annot=True)

    plt.figure(figsize=(10, 6))
    gb = df.groupby("NrSiblings").agg({"MathScore": 'mean', "ReadingScore": 'mean', "WritingScore": 'mean'})
    sns.heatmap(gb, annot=True)
    plt.title('Mean Scores by Sibling')
    plt.show()

    plt.figure(figsize=(10, 6))
    gb = df.groupby("LunchType").agg({"MathScore": 'mean', "ReadingScore": 'mean', "WritingScore": 'mean'})
    sns.heatmap(gb, annot=True)
    plt.title('Mean Scores by Food')
    plt.show()

    # Boxplot for mean with vertical line to mark mean
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='MathScore', data=df)
    plt.axvline(df['MathScore'].mean(), color='red', linestyle='dashed', linewidth=2, label='Mean')
    plt.title('Box Plot with Mean for Math Scores')
    plt.legend()
    plt.show()

    plt.figure(figsize=(8, 6))
    sns.boxplot(x='WritingScore', data=df)
    plt.axvline(df['WritingScore'].mean(), color='yellow', linestyle='dashed', linewidth=2, label='Mean')
    plt.title('Box Plot with Mean for Writing Scores')
    plt.legend()
    plt.show()

    plt.figure(figsize=(8, 6))
    sns.boxplot(x='ReadingScore', data=df)
    plt.axvline(df['ReadingScore'].mean(), color='red', linestyle='dashed', linewidth=2, label='Mean')
    plt.title('Box Plot with Mean for Reading Scores')
    plt.legend()
    plt.show()

# Function to calculate mean and standard deviation
def calculate_statistics():
    average_math = df['MathScore'].mean()
    average_reading = df['ReadingScore'].mean()
    average_writing = df['WritingScore'].mean()
    std_math = df['MathScore'].std()
    std_reading = df['ReadingScore'].std()
    std_writing = df['WritingScore'].std()

    return {
        'Mean of Math': average_math,
        'Mean of Reading': average_reading,
        'Mean of Writing': average_writing,
        'Std Dev of Math': std_math,
        'Std Dev of Reading': std_reading,
        'Std Dev of Writing': std_writing
    }

# Function to add student grade
def add_grade(df, Names, Gender, MathScore, ReadingScore, WritingScore, EthnicGroup, ParentEduc, LunchType, TestPrep,
              ParentMaritalStatus, PracticeSport, IsFirstChild, NrSiblings, TransportMeans, WklyStudyHours):
    new_entry = {
        'Names': Names,
        'Gender': Gender,
        'MathScore': int(MathScore),
        'ReadingScore': int(ReadingScore),
        'WritingScore': int(WritingScore),
        'EthnicGroup': EthnicGroup,
        'ParentEduc': ParentEduc,
        'LunchType': LunchType,
        'TestPrep': TestPrep,
        'ParentMaritalStatus': ParentMaritalStatus,
        'PracticeSport': PracticeSport,
        'IsFirstChild': IsFirstChild,
        'NrSiblings': NrSiblings,
        'TransportMeans': TransportMeans,
        'WklyStudyHours': WklyStudyHours
    }
    df = df.append(new_entry, ignore_index=True)
    file_path = "new data2.csv"
    df.to_csv(file_path, index=False)
    print("Successfully added.")
    return df

# Function to access specific student scores by entering name
def get_student_score(student_name):
    student_data = df[df['Names'] == student_name]

    math = student_data['MathScore'].item()
    reading = student_data['ReadingScore'].item()
    writing = student_data['WritingScore'].item()

    scores = [math, reading, writing]
    subjects = ["math", "reading", "writing"]

    plt.figure(figsize=(10, 6))
    ax = plt.bar(subjects, scores, color=['skyblue', 'orange', 'yellow'])
    plt.bar_label(ax)
    plt.show()
    return

# Assuming you have already loaded or initialized your DataFrame (df)
df = pd.read_csv("new data2.csv")  # Update with your actual file path

# Main loop for user interaction
while True:
    print("\nOptions:")
    print("1. Show Statistics")
    print("2. Visualize Data")
    print("3. Add Grade")
    print("4. Student Scores")
    print("5. Exit")
    choice = input("Enter your choice (1/2/3/4): ")
    if choice == '1':
        print(calculate_statistics())
    elif choice == '2':
        visualize_data()
    elif choice == '3':
        Names = input("Enter the Name: ")
        Gender = input("Enter the Gender: ")
        MathScore = input("Enter the Math score: ")
        ReadingScore = input("Enter the Reading score: ")
        WritingScore = input("Enter the Writing score: ")
        EthnicGroup = input("Enter the Ethnic Group: ")
        ParentEduc = input("Enter the parent education: ")
        LunchType = input("Enter the lunch type of student: ")
        TestPrep = input("Does the student did prep: ")
        ParentMaritalStatus = input("Enter the Parent Marital Status: ")
        PracticeSport = input("Does the student play sport: ")
        IsFirstChild = input("Is the student the first child: ")
        NrSiblings = input("Enter the number of siblings: ")
        TransportMeans = input("Enter the Transport mean of student: ")
        WklyStudyHours = input("Enter the weekly study hours: ")
        add_grade(df, Names, Gender, MathScore, ReadingScore, WritingScore, EthnicGroup, ParentEduc, LunchType, TestPrep,
                  ParentMaritalStatus, PracticeSport, IsFirstChild, NrSiblings, TransportMeans, WklyStudyHours)
    elif choice == '4':
        student_name = input("Enter the student name for student data: ")
        get_student_score(student_name)
    elif choice == '5':
        break
    else:
        print("Invalid option")
```
